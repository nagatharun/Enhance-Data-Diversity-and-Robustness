```python
import shutil
from ultralytics import YOLO
from pathlib import Path

# Step 1: Load your trained YOLOv8 model
model_path = '/student/vcheruku/Enhance-Data-Diversity-and-Robustness/yolo-for-single-class/YOLOv8/yolov8-single-object-detection/weights/best.pt'
model = YOLO(model_path)

# Step 2: Export the model to TFLite format with INT8 quantization
export_result = model.export(format='tflite', int8=True)

# Step 3: Move the exported file to your desired location
# export_result is a dictionary that includes 'filepath'
tflite_path = Path(export_result['filepath'])  # This is the full path to the exported TFLite model

# Define the destination path
destination_path = Path('/student/vcheruku/Enhance-Data-Diversity-and-Robustness/yolo-for-single-class/YOLOv8/yolov8-single-object-detection/weights/quantization/best_int8.tflite')
destination_path.parent.mkdir(parents=True, exist_ok=True)  # Ensure the directory exists

# Move the file
shutil.move(str(tflite_path), str(destination_path))

print(f"Quantized TFLite model exported to: {destination_path}")
```
